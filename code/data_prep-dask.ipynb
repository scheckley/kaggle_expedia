```python
import pandas as pd
pd.set_option('display.max_columns', 100)
import numpy as np
import dask.dataframe as dd
from tqdm import tnrange, tqdm_notebook as tqdm
```

```python
dat = dd.read_csv('../data/train.csv')
```

```python
len(dat[dat['is_booking']==1]), len(dat[dat['is_booking']==0])
```

```python
# remove non-bookings
dat = dat[dat['is_booking']!=0]
```

```python
dat.head()
```

```python
grouped = dat.groupby('user_id').is_booking.count().reset_index()
```

```python
grouped.head()
```

```python
agent_ids = grouped[grouped['is_booking']<20]
l = list(agent_ids['user_id'])
```

```python
len(agent_ids)
```

```python
ddf_selected = dat[dat['user_id'].isin(l)]
```
